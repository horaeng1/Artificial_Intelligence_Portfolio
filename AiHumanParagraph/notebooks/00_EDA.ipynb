```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DATA_PATH = ""
# ✅ CSV 불러오기 (문서 수준 or 문단 수준 중 하나 선택)
df_feat = pd.read_csv(DATA_PATH)
```

```python
# 📉 상위 1% 이상치 제거 함수
def clip_top_percentile(df, cols, percentile=99):
    df_clipped = df.copy()
    for col in cols:
        threshold = df[col].quantile(percentile / 100)
        df_clipped = df_clipped[df_clipped[col] <= threshold]
        print(f"{col} clipped above {threshold:.2f} (top {100 - percentile}% removed)")
    return df_clipped

num_cols = [
    "char_len",
    "sent_count",
    "avg_sent_len_char",
    "tok_len",
    "avg_sent_len_tok",
    "punct_ratio"
]

# ✂️ 상위 1% 제거
df_clipped = clip_top_percentile(df_feat, num_cols, percentile=99)

# 📊 제거 후 다시 시각화
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

for col in num_cols:
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f"Feature Comparison (clipped top 1%): {col}", fontsize=16, weight='bold')

    # Boxplot
    sns.boxplot(data=df_clipped, x="generated", y=col, palette="coolwarm", showfliers=False, ax=axes[0])
    axes[0].set_title("Boxplot (Clipped)")
    axes[0].set_xlabel("generated (0=Human, 1=AI)")
    axes[0].set_ylabel(col)

    # Violin
    sns.violinplot(data=df_clipped, x="generated", y=col, palette="coolwarm", inner="quartile", ax=axes[1])
    axes[1].set_title("Violin Plot (Clipped)")
    axes[1].set_xlabel("generated")
    axes[1].set_ylabel(col)

    # KDE
    sns.kdeplot(data=df_clipped[df_clipped["generated"] == 0][col], label="Human (0)", shade=True, ax=axes[2])
    sns.kdeplot(data=df_clipped[df_clipped["generated"] == 1][col], label="AI (1)", shade=True, ax=axes[2])
    axes[2].set_title("KDE (Clipped)")
    axes[2].set_xlabel(col)
    axes[2].set_ylabel("Density")
    axes[2].legend()

    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind, ks_2samp

# ✅ KS-stat 계산 함수 (아까 쓴 것 다시 사용)
def feature_significance(df, cols, label_col="generated"):
    results = []
    for col in cols:
        human = df[df[label_col] == 0][col].dropna()
        ai = df[df[label_col] == 1][col].dropna()
        t_stat, t_p = ttest_ind(human, ai, equal_var=False)
        ks_stat, ks_p = ks_2samp(human, ai)
        results.append({
            "feature": col,
            "human_mean": human.mean(),
            "ai_mean": ai.mean(),
            "t_p_value": t_p,
            "ks_stat": ks_stat,
            "ks_p_value": ks_p
        })
    return pd.DataFrame(results).sort_values("ks_stat", ascending=False)

stats_after = feature_significance(df_clipped, num_cols)

# ✅ KS 통계량 기준으로 중요도 시각화
plt.figure(figsize=(10, 6))
sns.barplot(data=stats_after, x="ks_stat", y="feature", palette="viridis")

plt.title("Feature Importance (KS-stat after top 1% removal)", fontsize=15, weight='bold')
plt.xlabel("KS Statistic (higher = stronger difference)")
plt.ylabel("Feature")
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.show()
```

```python
# 📈 결과 계산 (Before & After 비교)
stats_before = feature_significance(df_feat, num_cols)
stats_after = feature_significance(df_clipped, num_cols)

print("📊 Feature Significance – BEFORE Clipping")
display(stats_before)

print("📊 Feature Significance – AFTER Top 1% Clipping")
display(stats_after)
```

```python
from typing import Counter
def get_top_tokens(text, top_n=30):
  counter = Counter()
  tokens = []
  for paragraph in text:
    tokens = paragraph.split(" ")
    counter.update(tokens)
  return counter.most_common(top_n)

human_tokens = df_feat[df_feat["generated"] == 0]["tokens_preview"].tolist()
ai_tokens = df_feat[df_feat["generated"] == 1]["tokens_preview"].tolist()
```

```python
top_human_tokens = get_top_tokens(human_tokens)
top_ai_tokens = get_top_tokens(ai_tokens)
```

```python
def lexical_diversity(texts):
    all_tokens = []
    tokens = []
    for paragraph in texts:
      tokens = paragraph.split(" ")
      all_tokens.extend(tokens)
    total = len(all_tokens)
    unique = len(set(all_tokens))
    hapax = sum(1 for c in Counter(all_tokens).values() if c == 1)
    return {
        "total_tokens": total,
        "unique_tokens": unique,
        "TTR": unique / total,
        "hapax_ratio": hapax / total
    }

print("👤 Human diversity:", lexical_diversity(human_tokens))
print("🤖 AI diversity:", lexical_diversity(ai_tokens))
```
